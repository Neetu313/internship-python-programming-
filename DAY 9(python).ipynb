```python
import numpy  as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,4*np.pi, 100)# domain
Y = np.cos(X)               # range
plt.scatter(X,Y,marker = 'x', color = '#FF2C55')
plt.show()
```

```python
from ipywidgets import interact, interactive, fixed , interact_manual


X = np.linspace(0,4*np.pi, 100)# domain
Y = np.cos(X)               # range

def plotting(Ax):
    plt.plot(X,Y, color = '#FF2C55', ls = '-', linewidth = 1)
    plt.plot([Ax, Ax], [np.sin(Ax), 0], alpha = 0.3, color = 'black') # horizontal line
    plt.plot([Ax, 0], [np.sin(Ax), np.sin(Ax)] ,alpha = 0.4, color = 'black')  # vertical line
    plt.scatter(Ax,np.sin(Ax), color = 'green') 
    plt.text(Ax-0.35, np.sin(Ax)+0.04, 'p.o.i', style = 'italic')
    plt.axhline()
    plt.axvline()
    plt.show()

interact(plotting, Ax = np.pi/2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-4, 4, 200)
y = x**2 
plt.plot(x, y, color = 'red')
plt.axhline()
plt.axvline()
plt.show()
```
