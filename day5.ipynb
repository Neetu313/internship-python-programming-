# Matplotlib.pyplot

```python
import matplotlib.pyplot as plt
import numpy as np
X = []
for i in np.linspace(1,5,1000):        # linspace gives linear spaced values(equaly spaced) b/w two points like(1,3,100)
    X.append(i)
Y = []
for i in X:
    val = i*i
    Y.append(val)
plt.plot(X,Y)
plt.show()
```

```python
# another easy mathode
import matplotlib.pyplot as plt
import numpy as np
X = [sm_values for sm_values in np.linspace(1,4,100)]
Y = [values**2 for values in X]
plt.plot(X,Y)
Y_ = [1,4,9,16]
X_ = [1,2,3,4]   
plt.scatter(Y_,X_)   # scatter gives dots graph
plt.show() 
```
